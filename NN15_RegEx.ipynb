# Python RegEx

## RegEx Package

```python
# Busca en el string si empieza con "The" y termina con "Spain"
import re

txt = "The rain in Spain"
x = re.search("^The.*Spain$", txt)
print(x)
```

## La función findall( )

```python
# Imprime una lista con todas las coincidencias
import re

txt = "The rain in Spain"
x = re.findall("ai", txt)
print(x)
```

```python
# Retorna una lista vacia porque no hay coincidencias
import re

txt = "The rain in Spain"
x = re.findall("Portugal", txt)
print(x)
```

## La función search( )

```python
# Busca el primer carácter de espacio en el string
import re

txt = "The rain in Spain"
x = re.search("\s", txt)

print("The first white-space character is located in position:", x.start())
```

```python
# Retorna None en vez de un objeto match al no haber coincidencia
import re

txt = "The rain in Spain"
x = re.search("Portugal", txt)
print(x)
```

## La función split( )

```python
# Divide en una lista cada vez que encuentra el patron en el string
import re

txt = "The rain in Spain"
x = re.split("\s", txt)
print(x)
```

```python
# Divide al string solo en la primer coincidencia
import re

txt = "The rain in Spain"
x = re.split("\s", txt, 1)
print(x)
```

## La función sub( )

```python
# Reemplaza cada espacio por un _
import re

txt = "The rain in Spain"
x = re.sub("\s", "_", txt)
print(x)
```

```python
# Reemplaza los primeros 2 espacios por _
import re

txt = "The rain in Spain"
x = re.sub("\s", "_", txt, 2)
print(x)
```

# Match Object

## Contiene información de la busqueda y el resultado

## Si no hay coincidencias en vez de un objeto se retorna None

```python
# Imprime un objeto match

import re

txt = "The rain in Spain"
x = re.search("ai", txt)
print(x) 
```

```python
# Imprime la posición de la coincidencia

import re

txt = "The rain in Spain"
x = re.search(r"\bS\w+", txt)
print(x.span())
```

```python
# Imprime el string que tomo como base

import re

txt = "The rain in Spain"
x = re.search(r"\bS\w+", txt)
print(x.string)
```

```python
# Imprime la parte del string de la coincidencia

import re

txt = "The rain in Spain"
x = re.search(r"\bS\w+", txt)
print(x.group())
```
