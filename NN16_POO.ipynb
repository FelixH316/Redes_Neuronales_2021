## El constructor en Python

```python
# El método __init__ o mejor dicho el constructor se llama cada vez
# que se crea un objeto

class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

p1 = Person("John", 36)

print(p1.name)
print(p1.age)
```

## Métodos de la clase

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def myfunc(self):
    print("Hello my name is " + self.name)

p1 = Person("John", 36)
p1.myfunc()
```

## El parámetro self

```python
# self es el parametro que se usa como referencia a la instancia actual de la clase
# No necesariamente se tiene que llamar self, pero debe ser el primer parametro del método

class Person:
  def __init__(mysillyobject, name, age):
    mysillyobject.name = name
    mysillyobject.age = age

  def myfunc(abc):
    print("Hello my name is " + abc.name)

p1 = Person("John", 36)
p1.myfunc()
```

# Herencia en Python

## Creación de la clase padre

```python
class Person:
  def __init__(self, fname, lname):
    self.firstname = fname
    self.lastname = lname

  def printname(self):
    print(self.firstname, self.lastname)

#Use the Person class to create an object, and then execute the printname method:

x = Person("John", "Doe")
x.printname()
```

## Creación de la clase hija

```python
class Student(Person):
  pass
```

```python
x = Student("Mike", "Olsen")
x.printname()
```

## Constructor de la clase hija

```python
class Student(Person):
    def __init__(self, fname, lname):
        pass
        #add properties etc.
```

## Constructor de la clase hija llamando al constructor padre

```python
class Student(Person):
    def __init__(self, fname, lname, grade):
        Person.__init__(self, fname, lname)
        self.grade = grade
    
    def print_grade(self):
        print(self.grade)
    
x = Student("John", "Doe", 8)
x.print_grade()
```

## La Función super( )

```python
# super() hace que la clase hija herede todos los métodos y atributos del padre

class Student(Person):
    def __init__(self, fname, lname):
        super().__init__(fname, lname)
```

## Adición de atributos

```python
class Student(Person):
    def __init__(self, fname, lname):
        super().__init__(fname, lname)
        self.graduationyear = 2019
```

```python
class Student(Person):
    def __init__(self, fname, lname, year):
        super().__init__(fname, lname)
        self.graduationyear = year

x = Student("Mike", "Olsen", 2019)
```

## Adición de métodos

```python
# Si se añade un método a la clase hija y es el mismo nombre que un método del padre
# entonces se sobreescribirá

class Student(Person):
    def __init__(self, fname, lname, year):
        super().__init__(fname, lname)
        self.graduationyear = year

    def welcome(self):
        print("Welcome", self.firstname, self.lastname, "to the class of", self.graduationyear)

student1 = Student("Erick", "Negrete", 2018)
student1.welcome()
```
